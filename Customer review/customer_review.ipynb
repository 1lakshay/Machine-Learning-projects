```python
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv("Restaurant_Reviews.tsv",delimiter="\t")
```

```python
data
```

```python
print(len(data))
```

```python
# preprocessing the text 
import re 
import nltk
from nltk.corpus import stopwords  
from nltk.stem.porter import PorterStemmer
```

```python
corpus=[]    # store the strings of every row from dataset 
for i in range(len(data)):
    review=re.sub('[^a-zA-Z]'," ",data["Review"][i])  # removing any char which is not the part os alphabets 
    review=review.lower()     # converting each to lower letter 
    review=review.split()   # breaking string into words 
    ps=PorterStemmer()    # stemming
    all_stopwords=stopwords.words("english")  # importing stopwords of english lang
    all_stopwords.remove('not')   # removing "not" from the list of stopwords 
    review=[ps.stem(word) for word in review if word not in set(all_stopwords)]  # performing stemming 
    review =' '.join(review)    # joing words again to form a string 
    
    corpus.append(review)   # appending preprocessed string 
    print(corpus)
```

```python
# creating vector 
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
cv=CountVectorizer(max_features=1566)
X=cv.fit_transform(corpus).toarray()  # creating vector 
y=data.iloc[:,-1].values
```

```python
print(len(X[0]))     # finding size of vector  
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred.shape)
```

```python
print(np.concatenate((y_test.reshape(-1,1),y_pred.reshape(-1,1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
