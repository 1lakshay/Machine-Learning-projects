Checking for the presence of Null values

```python
data.rename(columns={" BMI ":"BMI","Life expectancy ":"Life_Expectancy","Adult Mortality":"Adult_Mortality",
                   "infant deaths":"Infant_Deaths","percentage expenditure":"Percentage_Exp","Hepatitis B":"HepatitisB",
                  "Measles ":"Measles"," BMI ":"BMI","under-five deaths ":"Under_Five_Deaths","Diphtheria ":"Diphtheria",
                  " HIV/AIDS":"HIV/AIDS"," thinness  1-19 years":"thinness_1to19_years"," thinness 5-9 years":"thinness_5to9_years","Income composition of resources":"Income_Comp_Of_Resources",
                   "Total expenditure":"Tot_Exp"},inplace=True)
```

Handling missing values in target variable by removing the entire row where there is NUll values
as imputing the missing values with other values can lead to the loss of valuable data 

```python
data.dropna(subset = ['Life_Expectancy'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.info()
```

Imputing values with their mean values 

```python
columns= ['Tot_Exp','Diphtheria','Population','thinness_1to19_years','thinness_5to9_years','Income_Comp_Of_Resources',
         'Schooling','GDP']
```

```python
from sklearn.impute import SimpleImputer 
```

```python
imputer = SimpleImputer(strategy = "mean")
data[columns] = imputer.fit_transform(data[columns])
```

```python
data.isnull().sum()
```

Imputing values with the median values 

```python
column = ['Alcohol','HepatitisB','BMI','Polio']
```

```python
imputer = SimpleImputer(strategy ='median')
data[column] = imputer.fit_transform(data[column])
```

```python
data.isnull().sum()
```

```python
data.info()
```

Conversion of categorical value to numerica using map() function 

```python
data['Status'].unique()
```

```python
data['Status'] = data['Status'].map({"Developing":0,
                   "Developed":1})
```

```python
data['Status'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
data['Country'] = lb.fit_transform(data['Country'])
```

```python
data['Country']
```

```python
data.info()
```

```python
data.corr()
```

```python
data.drop('Country',axis=1,inplace=True)
```

```python
data
```

```python
data.info()
```

```python
print(len(data.columns))
```

```python
X = data.iloc[:, [i for i in range(len(data.columns)) if i!=2]].values
y = data.iloc[:, 2].values
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
model_linear = LinearRegression()
model_linear.fit(X_train,y_train)
```

```python
y_pred = model_linear.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)      # instance is created 

X_poly_train = poly_reg.fit_transform(X_train)
X_poly_test = poly_reg.fit_transform(X_test)
```

```python
# with degree = 4 -> -1.12....
# with degree = 3 -> 0.3...
# with degree = 2 -> 0.868141..
```

```python
print(X_poly_train.shape)
print(X_poly_test.shape)
```

```python
model_linear.fit(X_poly_train,y_train)
```

```python
y_pred_poly = model_linear.predict(X_poly_test)
```

```python
r2_score(y_test,y_pred_poly)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dt = DecisionTreeRegressor(random_state=0)
model_dt.fit(X_train,y_train)
```

```python
y_pred_dt = model_dt.predict(X_test)
```

```python
r2_score(y_test,y_pred_dt)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rf = RandomForestRegressor(n_estimators = 10, random_state=0)
model_rf.fit(X_train,y_train)
```

```python
y_pred_dt = model_rf.predict(X_test)
```

```python
r2_score(y_test,y_pred_dt)
```

# Out of all the models Random Forest have the performing Best 

# Performing K fold cross validation : For evaluating the perfromance by checking the perfromnace with differenet sets of test sets 

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model_rf, X=X_train, y=y_train, cv=10)
print("accuracy = {:.2f} %".format(accuracies.mean()*100))
print("standard deviation = {:.2f} %".format(accuracies.std()*100))
```

# grid search

```python
from sklearn.model_selection import GridSearchCV
parameters = {"n_estimators" :[10,50,100,150,200]}
grid = GridSearchCV(estimator = model_rf,
                   param_grid = parameters,
                   scoring="r2",          # metric 
                   cv=10,                 # no of bins 
                   n_jobs = -1            # no of processors 
                   )
grid.fit(X_train,y_train)
best_accuracy = grid.best_score_
best_parameter = grid.best_params_
print("best accuracy = {:.2f} %".format(best_accuracy*100))
print("best parameter = ",best_parameter)
# with n_estimator vaulues as [10,20,30,40,50,60,70]
# best accuracy got is 96.04%
# best parameter got is 70
```
