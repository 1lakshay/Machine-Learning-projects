```python
# splitting the dataset 
X = data.drop(columns = ['quality'],axis=1)
print(X.info())
```

# Labelling:
# can be done in 2 ways: Binary Encoding & integer Encoding 
# as here the focus is on finding the Good quality so we used Binary Encoding
# Integer Encoding: done only when we have to predict the all the classes

```python
# as in this model we are more focused on findign only the good quality so we are using the Binary encoding in which 
# 1 for good quality >=7
# 0 for poor qulity for remaining value 
```

```python
# helper function for Binary encoding
def limiter(value):
    if value>=7:
        return 1 
    else:
        return 0
```

```python
data['quality'] = data['quality'].apply(limiter)
```

```python
data['quality'].value_counts()
```

```python
Y=data['quality']
print(Y.info())
```

```python
# splitting of training and testing data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# the prediction is made from all the independent variables 
X_test_pred = model.predict(X_test)
```

```python
# all the ouput as 0 & 1 
print(X_test_pred)
```

```python
print(X_test_pred.shape)
```

```python
print(X_test.shape)
```

```python
# calculating the accuracy score with seeing in difference betwen predicted value and actual value
test_data_accuracy = accuracy_score(X_test_pred,Y_test)
print(test_data_accuracy)
```

# if we dont do the binary encoding the acuuracy we are getting is less than what we are getting now  which is 0.715625

# predictive system: For predicting the value on new and unseen data 

```python
# this is directly sending the value as 2D array 
# input_data_reshaped= [[7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0]]


input_data = (5.6,0.615,0.0,1.6,0.08900000000000001,16.0,59.0,0.9943,3.58,0.52,9.9)
# converting to array 
input_data_ass_arr = np.asarray(input_data)
print(input_data_ass_arr.shape)
print(input_data_ass_arr)
# reshaping the data 
input_data_reshaped = input_data_ass_arr.reshape(1,-1)
print(input_data_reshaped.shape)
print(input_data_reshaped)
```

```python
# the shape before reshaping array was (11,) means that it have only rows and 0 columns 
# so it will go row-wise but as we know that it is the values of one particular row 
# as we know that row corresponds to the sample and the column corresponds to the columns 
# that's why we reshaped it into (1,11) means it have 1 row and 11 columns 
# so it will go to the paritcular sample 
```

```python
def conversion(value):
    if value==1:
        return 'good'
    else:
        return 'bad'
```

```python
pred = model.predict(input_data_reshaped)
for i in pred:
    print(conversion(i))
```
